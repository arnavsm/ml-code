# Initializing Different Types of Array

```python
#All Zero Matrix
np.zeros((2,3))
```

```python
np.zeros((2,3), dtype = int)
```

```python
#All One Matrix
np.ones((4,2,3,4)) #You can use dtype as well
```

```python
#Any other number
np.full((2,2), 99, dtype = int)
```

```python
np.full(a.shape, 7) #Any other number (full_like) which is equal to (below)
```

```python
np.full_like(a, 4)
```

```python
np.arange(0,100).reshape(10,2,5)
```

```python
# Random
np.random.rand(4,2) #You don't insert a tuple but just directly. Difficult to remember but important
```

```python
np.random.randint(-4, 8, size = (3,3))
```

```python
#Identity Matrix
np.identity(4)
```

```python
arr = np.array([1,2,3])
r1 = np.repeat(arr, 3, axis = 0)
print(r1)
```

```python
arr = np.array([[1,2,3]])
r1 = np.repeat(arr, 3, axis = 0)
print(r1)
```

```python
#Inserting array into array
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
z[1,1] = 7
print(z)

output[1:4,1:4] = z
print(output)
```

```python
#Important!!! We aren't copying the arrays here

a = np.array([1,2,3])
b = a
b[0] = 100

print(b)
print(a)
```

```python
#Solution- copy() function
a = np.array([1,2,3])
b = a.copy()
b[0] = 100

print(b)
print(a)
```

# Reorganising Arrays

```python
#Stacking Arrays/Vectors
v1 = np.array([[1,2,3,4],[5,6,7,8]])
v2 = np.array([[3,45,56,76], [300,301,302,303]])
np.vstack([v1,v2,v1]) #Square Brackets or Round (any)
```
